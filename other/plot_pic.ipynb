```python
import matplotlib.pyplot as plt

# 准备数据
x = [1, 2, 3, 4, 5]
y1 = [8.2, 8, 9, 7.6, 8]
y2 = [2.2, 2.3, 1.9, 1.9, 1.7]
y3 = [2.2, 1, 1.1, 1.8, 2.2]
# 以下是设置中文字体的常见方法
plt.rcParams['font.family'] = 'Songti SC'  # 例如设置为宋体
# plt.rcParams['font.family'] = 'Microsoft YaHei'
# 创建一个图形
plt.figure()

# 绘制第一条折线，并设置节点形状
plt.plot(x, y1, marker='o', label='单词不懂')

# 绘制第二条折线，并设置节点形状
plt.plot(x, y2, marker='s', label='句子不懂')

# 绘制第三条折线，并设置节点形状
plt.plot(x, y3, marker='*', label='走神')

# 添加标题和坐标轴标签
# plt.title('Multiple Line Plots')
plt.xlabel('感兴趣程度')
plt.ylabel('标注数量')

# 扩大 y 轴范围
plt.ylim(0, 12)

# 显示图例在右上角
plt.legend(loc='upper right')

# 显示图形
plt.show()
```

```python
import matplotlib.pyplot as plt

words = ['P1', 'P2', 'P3', 'P4', 'P5', 'P6', 'P7', 'P8', 'P9', 'P10', 'P11', 'P12', 'P13']
values1 = [0.6, 0.62, 0.7, 1.0, 1.0, 1.0, 0.6, 1.0, 1.0, 0.6, 0.5, 0.26, 0.64]
values2 = [0.7, 0.4, 1.0,  1.0, 1.0, 0.7, 0.9, 1.0, 1.0, 1.0, 0.0, 0.30, 0.2]  # 假设的第二列数据
plt.rcParams['font.family'] = 'Songti SC'  # 例如设置为宋体

x = range(len(words))
width = 0.35  # 柱子的宽度

plt.figure(figsize=(10, 6))  # 增加图形大小，例如宽度为 10，高度为 6

plt.bar([i - width/2 for i in x], values1, width, label='单词')
plt.bar([i + width/2 for i in x], values2, width, label='句子')

plt.xlabel('')
plt.ylabel('阅读增益')
plt.title('')
plt.xticks(x, words)  # 设置 x 轴刻度标签为 words
plt.legend()  # 显示图例
plt.show()
```

```python
import matplotlib.pyplot as plt

words = ['P1', 'P2', 'P3', 'P4', 'P5', 'P6', 'P7', 'P8', 'P9', 'P10', 'P11*', 'P12*', 'P13*']
values1 = [1.0, 0.84, 0.9, 0.8, 0.82, 0.78, 0.9, 0.84, 0.8, 0.82, 0.42, 0.7, 0.3]
values2 = [0.0, 0.16, 0.1, 0.2, 0.18, 0.22, 0.1, 0.16, 0.2, 0.18, 0.58, 0.3, 0.7]  # 假设的第二列数据
plt.rcParams['font.family'] = 'Songti SC'  # 例如设置为宋体

x = range(len(words))
width = 0.35  # 柱子的宽度

plt.figure(figsize=(12, 6))  # 增加图形大小，例如宽度为 10，高度为 6

plt.bar([i - width/2 for i in x], values1, width, label='观察 I')
plt.bar([i + width/2 for i in x], values2, width, label='观察 II & III')

plt.xlabel('')
plt.ylabel('比例')
plt.title('')
plt.xticks(x, words)  # 设置 x 轴刻度标签为 words
plt.legend()  # 显示图例
plt.show()
```
